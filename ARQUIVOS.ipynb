# Abrir Arquivos

```python
arquivo = open('Machado_de_Assis.txt', 'r')
```

```python
conteudo = arquivo.readlines()
arquivo.close()
```

# Imprimir linha a linha

```python

for linha in conteudo:
  print(linha)
```

# Substituir uma palavra em todo o texto

```python
x = ''
for linha in conteudo:
  if linha != '':
    x += linha


print(x)
x = x.replace('Machado', 'Axe')
print('_______________')
print(x)
```

# Gravar um novo arquivo

```python
arquivo_novo = open('Axe_de_Assis.txt', 'w')
arquivo_novo.writelines(x)
arquivo_novo.close()
```

```python
arquivo_novo = open('Axe_de_Assis.txt', 'r')
arquivo_novo.read()
```

```python
arquivo_novo.close()
```

# Trabalhando com arquivos CSV

```python
import csv
with open('Tesla.csv', 'r') as f:
  leitor = csv.reader(f, delimiter = ',')
  for linha in leitor:
    data = linha[0]
    fechamento = linha[4]
    volume = linha[5]
    print('Data: ',data, '-__Fechamento : ', fechamento, '__Volume: ', volume)
```
