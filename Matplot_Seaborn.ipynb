Exemplo aplicado a saúde

```python
dias = np.arange(1,31)
dias
```

```python
vacinados = np.random.randint(0, 1000, 30)
contagios = np.random.randint(0, 700, 30)
vacinados
contagios
```

Cores nos gráficos

```python
#plt.style.use('classisc')
plt.style.use('dark_background')
#plt.style.use('default')

plt.figure(figsize=(10,5))
plt.bar(dias, vacinados)
# plt.plot(dias, vacinados, 'r')
plt.xlabel(' Dias do Mês ')
plt.ylabel(' Vacinados por Dia')
plt.show()
```

# Gráfico a partir de um DATAFRAME


```python
dados = pd.DataFrame (dias, columns= ['Dias'])
```

```python
dados['Contagios'] = contagios
dados['Vacinados'] = vacinados
```

```python
dados
```

```python
dados.plot(kind='bar', x='Contagios', y='Vacinados')
plt.ylabel('Vacinados')
plt.show()
```

# Seaborn

```python
sns.barplot(data=dados, x='Contagios', y='Vacinados')
plt.style.use('default')

```

```python
sns.lineplot(data=dados, x='Contagios', y='Vacinados')
```
