# Apagar linhas em que o atrobuto é nulo

```python
dados.dropna(subset=['loser_rank'], inplace=True)
```

# Apagar coluna em que todos os valores são nulos

```python
dados.dropna(axis=1, how='all')
```

# Alterar valor dados nulos de uma coluna

```python
dados['loser_entry'].unique()
```

# Alterar mais de um atrobuto com condições diferentes

```python
dados['w_ace'].unique()
```

```python
dados['loser_ht'].unique()
```

```python
dados.fillna(value={'w_ace':0, 'loser_ht':dados['loser_ht'].mean()}, inplace=True)
```

```python
dados['loser_ht'].unique()
```

# Metodo ffill, projeta dado anterios ao proximo nulo

```python
dados['minutes'][1975:1990]
```

```python
dados['minutes'].fillna(method='ffill', inplace=True)
```

```python
dados['minutes'][1975:1990]
```
