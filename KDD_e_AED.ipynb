# Matriz de correlação (Pearson)

quando proximo de 1 possui correlção, quando proximo de 0 nenhuma relação e quan proximo -1 correlação negativa.

```python
mask = np.triu(np.ones_like(dados.corr(), dtype=np.bool))
```

```python
plt.figure(figsize=(20,20))
sns.heatmap(dados.corr(), mask = mask, square = True, annot=True, vmin=-1, vmax=1)
plt.show()
```
